```python
import pandas as pd
import matplotlib.pyplot as plt
import psycopg2
from sqlalchemy import create_engine
from postgres import password
```

```python
# Create the connection
engine = create_engine(f'postgresql://postgres:{password}@localhost:5432/Employees')
conn = engine.connect()
```

```python
# Read salary table into a DataFrame
salary = pd.read_sql("Select * From Salaries",conn)
salary.head()
```

```python
# Create histogram to visualize salary ranges
plt.hist(salary.salary)
plt.title("Salary Distribution")
plt.xlabel("Salary")
plt.ylabel("Number of Employees")
plt.show()
```

```python
# Read title table into a DataFrame
title = pd.read_sql("Select * From titles", conn)
title.head()
```

```python
# Read employee table into a DataFrame
employee = pd.read_sql("Select * From employees", conn)
employee.head()
```

```python
# Merge employee and salary dfs
employee_salary = pd.merge(employee,salary,on="emp_no", how="inner")
employee_salary.head()
```

```python
# Merge title df into employee_salary
complete_info = pd.merge(employee_salary,title,left_on="emp_title_id",right_on="title_id",how="outer")
complete_info.head()
```

```python
# Average salary by title
salary_title = complete_info.loc[:,["salary","title"]]
salarys_by_title = salary_title.groupby("title").mean()
salarys_by_title
```

```python
# Create bar chart
salary_bar = salarys_by_title.plot(kind='bar',rot=45)
salary_bar.set_xlabel("Job Title")
salary_bar.set_ylabel("Average Salary")
salary_bar.set_title("Average Salary by Job Title")
plt.show()
```
